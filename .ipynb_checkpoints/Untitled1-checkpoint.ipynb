```python
df = pd.get_dummies(data=df, columns=[ 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection','Contract','PaymentMethod','PaperlessBilling','Churn'])
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.head()
```

```python
df.columns
```

```python
x = df[['tenure', 'InternetService_DSL','InternetService_Fiber optic', 'OnlineSecurity_Yes',
       'Contract_Month-to-month', 'PaperlessBilling_Yes', 'PaymentMethod_Electronic check','OnlineBackup_Yes']]
y = df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
x_train.head()
```

```python
x_train.describe()
```

```python
sns.boxplot(x=x_train["tenure"])
```

<blockquote>
Looking at the boxplot there don't seem to  be any extreme or incorrect values.    
</blockquote>

```python
for 
```

```python
dt = DecisionTreeClassifier(max_depth=5)
dt.fit(x_train, y_train)
```

```python
dot_data = StringIO()
export_graphviz(dt, out_file=dot_data, feature_names=x_train.columns,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png())
```

```python
import graphviz 
dot_data = tree.export_graphviz(dt)  
graph = graphviz.Source(dot_data)  
graph.render("iris")
```

```python
y_pred = dt.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
